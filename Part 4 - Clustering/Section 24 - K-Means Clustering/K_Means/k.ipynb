```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset  = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters =i, init = "k-means++", max_iter =300, n_init =10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11) , wcss)
#wcss
```

```python
plt.show()
```

```python
kmeans = KMeans(n_clusters =5, init = "k-means++", max_iter =300, n_init =10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)

```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100, c='red',label='1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100, c='blue',label='2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100, c='yellow',label='3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100, c='cyan',label='4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],s=100, c='green',label='5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c="yellow", label="centroid")
plt.show()
```
